```python
models_parameters = pd.DataFrame( columns=['regression_model', 'parameter_list', 'no_of_params'])
```

```python
from sklearn.linear_model import Ridge
from sklearn.feature_selection import RFECV
```

```python
model_names = {  
    'linear_regression':  LinearRegression(),
    'ridge_regression' : Ridge(alpha=0.2, random_state=0),
    'xg_boost': XGBRegressor() ,
    'lg_boost': LGBMRegressor()
    }
```

```python

for model_name, mp in model_names.items():
    to_use = []
    name = model_name
    estimator = mp
    # Use RFE with cross-validation to 
    # find the optimal number of features 
    selector = RFECV(estimator, cv=5, min_features_to_select= 5, scoring = 'neg_mean_absolute_error') 
    selector = selector.fit(master_table, y) 

    # Print the optimal number of features 
    print("Optimal number of features: %d" % selector.n_features_) 

    # Print the selected features 
    #print("Selected features: %s" % selector.support_) 
    
    lis = np.array(columns)
    
    filter = np.array(selector.support_)
    #to_use.append(lis[filter])
    to_use = (lis[filter].tolist()) 
    no_of_params = len(to_use)
    print(to_use)
    models_parameters.loc[len(models_parameters.index)] = [model_name, to_use, no_of_params]  
```
