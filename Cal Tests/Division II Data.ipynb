```python
import numpy as np
import pandas as pd
import glob
```

```python
files = glob.glob('*.csv')
for file in files:
    print(file)
```

```python
file_dict = {}
col_dict = {}
columns = []
for file in files:
    data_2024 = pd.read_csv(file)
    # data_2024 = data_2024.dropna(thresh=4)
    # #data_2024 = data_2024.dropna(data_2024[(data_2024["Header 1"]) & (data_2024["Header 2"])])
    # data_2024 = data_2024.reset_index()
    # data_2024
    #data_2024.iloc[7104]['Header 1']
    frames_dict = {}

    for i in range(len(data_2024)):
        t = pd.DataFrame(data_2024.loc[[i]], columns=['Header 1', 'Header 2', 'Header 3', 'Header 4', 'Header 5', 'Header 6', 'Header 7', 'Header 8', 'Header 9'])
        if t.iloc[0]['Header 1'] == "Rank":
            name = i
            frames_dict[name]= []
        #print(name, i)
        # try:
        #     print(frames_dict[7104])
        # except:
        #     continue
        frames_dict[name].append(pd.DataFrame(data_2024.loc[[i]]))
        #print(frames_dict.keys())

    #frames_dict[0][0]
    frames_dict_2 = {}
    for key in frames_dict.keys():
        frames_dict_2[key] = pd.DataFrame(frames_dict[key][0])
        for i in range(len(frames_dict[key])):
            if i == 0:
                continue
            else:
                frames_dict_2[key] = pd.concat([frames_dict_2[key], frames_dict[key][i]], sort=False)

    frames_dict_2[0]
    for a in frames_dict_2.keys():
        frames_dict_2[a].columns = frames_dict_2[a].iloc[0]
        frames_dict_2[a] = frames_dict_2[a].drop([a])
        frames_dict_2[a] = frames_dict_2[a].dropna(axis=1, how='all')
        frames_dict_2[a] = frames_dict_2[a].drop(columns = ['Rank'])
        
        if "W-L" in frames_dict_2[a].columns:
            frames_dict_2[a] = frames_dict_2[a].drop(columns = ["W-L"])
        if "GM" in frames_dict_2[a].columns:
            frames_dict_2[a] = frames_dict_2[a].drop(columns = ["GM"])
        if "Avg" in frames_dict_2[a].columns:
            frames_dict_2[a] = frames_dict_2[a].drop(columns = ["Avg"])
        if "REB" in frames_dict_2[a].columns:
            frames_dict_2[a] = frames_dict_2[a].drop(columns = ["REB"])
        if "RPG" in frames_dict_2[a].columns:
            frames_dict_2[a] = frames_dict_2[a].drop(columns = ["RPG"])
        if "Ratio" in frames_dict_2[a].columns:
            frames_dict_2[a] = frames_dict_2[a].drop(columns = ["Ratio"])
        if "Pct" in frames_dict_2[a].columns:
            frames_dict_2[a] = frames_dict_2[a].drop(columns = ["Pct"])
        if "PPG" in frames_dict_2[a].columns:
            frames_dict_2[a] = frames_dict_2[a].drop(columns = ["PPG"])
        if "3FG" in frames_dict_2[a].columns:
            frames_dict_2[a] = frames_dict_2[a].drop(columns = ["3FG"])
        frames_dict_2[a] = frames_dict_2[a].loc[:,~frames_dict_2[a].columns.duplicated()]

        frames_dict_2[a] = frames_dict_2[a].dropna(how="all")
        
        #print(len(frames_dict_2[a].columns))
        # for column in frames_dict_2[a].columns:
        #     column = column.astype(str)
        #     print(column)
        #     if "nan" in frames_dict_2[a].columns:
        #         frames_dict_2[a] = frames_dict_2[a].drop(columns=column)
    #print(type(column))
    #frames_dict_2[4144]


    #print(frames_dict_2.keys())
    year_df = frames_dict_2[0]

    for keya in frames_dict_2.keys():
        #print(keya)
        if keya == 0:
            continue
        else:
            #print(keya)
            year_df = pd.merge(year_df, frames_dict_2[keya], on="Name", how="left")
            #, suffixes=("_x", "_y")
    for col in year_df.columns:
        if "_y" in col:
            year_df = year_df.drop(columns=[col])
        if "_z" in col:
            year_df = year_df.drop(columns=[col])
        if "_x" in col:
            year_df = year_df.rename(columns={col: col[:-2]})
    year_df = year_df.loc[:,~year_df.columns.duplicated()]
    year_df['Name'] = year_df['Name'] + "_" + file[-8:-4]
    #print(len(year_df))
    year_df = year_df.reset_index(drop=True)
    #print(year_df.columns)
    file_dict[file] = year_df
    #print(file_dict[file].columns)
    col_dict[file] = file_dict[file].columns
    
for i in range(2):
    if i <1:
        cols1 = list(col_dict[file])
    else:
        cols2 = list(col_dict[file])
#print(cols1)
#print(cols2)

def common_member(a, b):
    # Return a list containing elements from list 'x' that also exist in list 'y'.
    return [item for item in a if item in b]

cols_to_keep = common_member(cols1, cols2)

count = 0
for file_name in file_dict.keys():
    temp = file_dict[file_name]
    temp = temp[list(cols_to_keep)]
    #print("FULL ", file_name)
    #display(temp)
    #print(count)
    if count == 0:
        final_df = temp
        count+=1
    else:
        #file_dict[file_name]= file_dict[file_name].loc[~file_dict[file_name].duplicated(keep='first')]
        #final_df = final_df.loc[~final_df.index.duplicated(keep='first')]
        #final_df = final_df.join(file_dict[file_name], how="outer")
        #display(file_dict[file_name])
        
        new_df = pd.concat([final_df, temp], sort=False)
        final_df = new_df
        
final_df
```
